```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
obj = pd.read_csv('data.csv')
obj
```

```python
inten = "Light Intensity".split()
ang = "Angle (rad)".split()
dropp = []
for elem in obj.columns:
    elem_lst = elem.split()
    if elem_lst[:2] != inten and elem_lst[:2] != ang:
        dropp.append(elem)
obj = obj.drop(dropp, axis=1)
obj
```

```python
y = obj['Light Intensity (%) Run #1'].values
x = obj["Angle (rad) Run #1"].values
for i in range(len(x)):
    x[i] = x[i]+43
plt.plot(x,y)
plt.title("The Sodium Spectrum")
plt.xlabel("Angle (rad)")
plt.ylabel("Light Intensity Percent")
```

```python
y = obj['Light Intensity (%) Run #2'].values
x = obj["Angle (rad) Run #2"].values
for i in range(len(x)):
    x[i] = x[i]-30
plt.plot(x,y)
plt.title("The Hydrogen Spectrum")
plt.xlabel("Angle (rad)")
plt.ylabel("Light Intensity Percent")
```

```python
y = obj['Light Intensity (%) Run #3'].values
x = obj["Angle (rad) Run #3"].values
for i in range(len(x)):
    x[i] = x[i]-1
plt.plot(x,y)
plt.title("The Helium Spectrum")
plt.xlabel("Angle (rad)")
plt.ylabel("Light Intensity Percent")
```
