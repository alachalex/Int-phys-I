```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
obj = pd.read_csv('1.csv')
```

```python
obj.head()
```

```python
#200 grams
x1 = obj["Time (s) Run #1"]
y1 = obj["Block-to-Block Times (s) Run #1"]
t1 = obj["Period (s) Run #1"]
avg_t = t1.mean(axis =0)
plt.plot(x1,y1,"ro")
plt.title("Time vs Block-to-Block Times")
plt.xlabel("Time (s)")
plt.ylabel("Block-to-Block Times")
print(avg_t)
```
